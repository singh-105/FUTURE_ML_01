```python
import pandas as pd
from prophet import Prophet
import os

# 1. Load data
df = pd.read_csv('../data/cleaned_sales_data.csv')

# 2. Prepare data
prophet_df = df[['data', 'venda']].rename(columns={'data': 'ds', 'venda': 'y'})

# 3. SUPER FAST MODEL 
model = Prophet(uncertainty_samples=0) # Removed seasonality to make it instant
model.fit(prophet_df)

# 4. Forecast 
future = model.make_future_dataframe(periods=90)
forecast = model.predict(future)

# 5. Force Save
output_path = '../data/forecast_results.csv'
forecast[['ds', 'yhat']].to_csv(output_path, index=False)

print(f"SUCCESS! The file is saved at: {os.path.abspath(output_path)}")
```
